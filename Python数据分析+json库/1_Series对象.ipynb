Seires对象

```python
import pandas as pd
import numpy as np
```

# 1.series的创建

```python
# 1.1 列表
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([1, 2, 3, 4, 5], index = ['a', 'b', 'c', 'd', 'e'], name = '数值')
print(s1)
print(s2)
```

```python
# 1.2 元组
s3 = pd.Series((1, 2, 3, 4, 5))
s4 = pd.Series((1, 2, 3, 4, 5), index = ['a', 'b', 'c', 'd', 'e'], name = '数值')
print(s3)
print(s4)
```

```python
# 1.3 字典
s5 = pd.Series({'a':1, 'b':2, 'c':3, 'd':4, 'e':5})
print(s5)
```

```python
# 1.4 numpy
s6 = pd.Series(np.array([1, 2, 3, 4, 5]))
s7 = pd.Series(np.arange(5))
s8 = pd.Series(np.arange(5), index = ['a', 'b', 'c', 'd', 'e'], name = '数值')
print(s6)
print(s7)
print(s8)
```

```python
# 1.5 data + index
s9 = pd.Series(data = [1, 2, 3, 4 ,5], index = ['a', 'b', 'c', 'd', 'e'], name = '数值')
print(s9)
```

2. Series的常用属性

```python
# 2.1 index 索引
print(s2.index)
```

```python
# 2.2 values 数据值
print(s2.values)
```

```python
# 2.3 使用索引搜索或更改值
print(s2['a'])   # 使用索引不可以直接将索引部分写出来（即s2['a']），要用print显示(即print(s2['a'])）
s2['a'] = 7
print(s2)
```
