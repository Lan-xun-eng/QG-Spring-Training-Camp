```python
import pandas as pd
import numpy as np
```

# 1.DataFrame的创建

```python
# 1.1 字典 + 列表
data1 = {
    'name':['张三', '李四', '王五'],
    'gender':['女', '男', '女'],
    'age':[11, 22, 33]
}
df1 = pd.DataFrame(data = data1)
df2 = pd.DataFrame(data = data1, index=['同学'+str(i) for i in range(3)])
print(df1)
print(df2)
```

```python
# 1.2 列表 + 元组
data2 = [
    ('张三', '女', 11),
    ('李四', '男', 22),
    ('王五', '女', 33)
]
df3 = pd.DataFrame(data = data2)
df4 = pd.DataFrame(data = data2, columns=['name', 'gender', 'age'], index=['同学'+str(i) for i in range(3)])
print(df3)
print(df4)
```

```python
# 1.3 numpy的ndarray数组
data3 = np.arange(12).reshape(3, 4)
df5 = pd.DataFrame(data = data3)
df6 = pd.DataFrame(data = data3, columns=['Math', 'Chinese', 'English', 'History'], index=['同学'+str(i) for i in range(3)])
print(df5)
df6
```

# 5.DataFrame的索引操作

--> 索引操作通常用来 获取元素数据 + 改变元素数据

！！！对更改数据，使用行列索引是不合理的，会有警告，往往使用loc/iloc

```python
df = pd.read_excel("E:\\pycharm项目\\data_analysis\\data\\US-pumpkins_data.xlsx")
index_name = ['案例'+str(i) for i in range(df.shape[0])]
df.index = index_name
df
```

```python
# 5.1 使用行列索引获取元素 -> 先列后行
df['City Name']['案例7']  # 只可以用来获取单个数据
```

```python
# 5.2 使用loc/iloc来获取数据/改变数据
# 5.2.1 loc[行索引，列名]
df.loc['案例8','City Name']  # 获取单个数据
df.loc['案例0':'案例5', 'City Name':'Package']  # 获取多个数据

df.loc['案例8','City Name'] = 'BEIJING'

# 5.2.2 iloc[行号，列编号]
df.iloc[0, 2]
df.iloc[0:4, 0:5]

df.iloc[0, 2] = '12'
df.iloc[0:4, 0:5]
```

```python
# 5.3 获取某一列的数据
df['Package']
df.Package
```

札记：
- 后者会更加方便，也更加快捷，但是，其有一个弊端：当列名中有空格时，这可以用第一种方法。即：对于Low Price，只可以用df['Low Price'],而不可以用df.Low Price

```python
# 5.4 获取多列数据
df[['Package', 'Low Price']]
```

札记：
- 要索引多列时，记得要用[]包裹表示一个整体

# 6.DataFrame的排序操作

# 6.DataFrame的基本运算

## 6.1 算数运算

```python
# 6.1.1 加法
df['Low Price'].add(2)
df['Low Price'] + 2
```

```python
# # 6.1.2 减法
df['Low Price'].sub(10)
df['Low Price'] - 10
```

## 6.2 逻辑运算 & | !

```python
# 6.2.1 普通方括号索引
df[df['Low Price'] > 150]
df[(df['Low Price'] >150) & (df['Low Price'] < 200)]
```

札记：如果有多个逻辑判断条件，则要在每个条件处加上小括号()

```python
# 6.2.2 query()
df.query('`Low Price` > 150')
df.query('`Low Price` > 150 & `Low Price` < 200')
```

札记:
- 对于query()函数，如果列名存在空格可以用反引号解决

```python
# 6.2.3 固定值筛选isin()
# df[(df['Low Price'] == 150) | (df['Low Price'] == 270)]
df.query('`Low Price` == 150 | `Low Price` == 270')
# df[df['Low Price'].isin([150, 270])]
```

注意：记清楚isin函数的使用细节

## 6.4 自定义运算

```python
# 6.4.1 apply()
# df对象.apply(自定义函数, axis = 0/1) -> 0表示列，1表示行
```

```python
# 分解版
def my_func(x):
    return x.max() - x.min()

df[['Low Price', 'High Price']].apply(my_func)
```

```python
# 分解版
def my_func(x):
    return x.max() - x.min()

df[['Low Price', 'High Price']].apply(my_func, axis = 0)
```

```python
# 完整版
df[['Low Price', 'High Price']].apply(lambda x: x.max() - x.min())
```
