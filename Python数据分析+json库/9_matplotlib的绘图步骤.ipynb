```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir(r'E:\\pycharm项目\\data_analysis')

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 1. 绘图步骤

1. 准备数据（准备x值和y值）
2. 创建画布
3. 绘制图像
4. 显示图像

```python
# 1.准备数据
x = range(60)
y = [np.random.uniform(15, 18) for i in x]
```

```python
# 2.创建画布
# figsize   dpi
# 参1：画布大小（宽，高）   参2：像素密度（正常是80~100）
plt.figure(figsize = (10,10), dpi = 100)
```

```python
# 3.绘制图像
# 参1：x值（x轴数据）   参2：y值（y轴数据）
plt.plot(x, y)

# 4.显示图像
plt.show()
```

# 2. 绘制图像的具体操作

1. x轴刻度及刻度标签 + y轴刻度及刻度标签
2. 网格
3. 标题
4. x轴标签 + y轴标签
5. 保存图像

**! ! !** **保存图像**一定要在**显示图像**前面，因为plt.show()还带有释放figure的内存的功能

## 2.1 刻度及刻度标签
- plt.xticks(x, **kwarg)
- plt.yticks(y, **kwarg)

    值得注意的是：原本绘制图像时，编译器会根据提供的x值和y值自动的生成x轴和y轴，
    但是我们可以利用xticks和yticks更改刻度将其换成我们想要的刻度。然而，刻度和刻度信息的替换要遵循以下顺序：

    原本的刻度 -> 替换后的刻度 -> 刻度标签

    所以，如果只是替换刻度则直接plt.xticks(x_ticks)即可，如果是要替换成刻度标签则要plt.xticks(x_tick， x_ticks_lable)

# 2.2 添加网格
- plt.grid(linestyle, alpha)
- 参1：网格线的格式   -  --  -.  :
- 参2：网格线的透明度(一般是0.3/0.5 ，默认是1)

## 2.3 添加标题
- plt.title(label, fontsize)
- 参1：标题内容
- 参2：字体大小

## 2.4 添加x轴/y轴标签
- plt.xlabel(xlabel, fontsize)
- 参1：x轴标签内容
- 参2：字体大小

- plt.ylabel(ylabel, fontsize)
- 参1：y轴标签内容
- 参2：字体大小

## 2.5 保存图像
- plt.savefig(保存路径)

```python
# 1.准备数据
x = range(0, 60)
y = [np.random.uniform(15, 28) for i in x]

# 2.创建画布
plt.figure(figsize = (10, 10), dpi = 100)

# 3.绘制图像
## 3.1 绘制初始折线图
plt.plot(x, y)

## 3.2 添加刻度及刻度信息
x_ticks = range(0, 60, 5)
y_ticks = range(0, 40, 5)
x_ticks_label = [f'11点{i}分' for i in x_ticks]

plt.xticks(x_ticks, x_ticks_label)
plt.yticks(y_ticks)

## 3.3 添加网格
plt.grid(linestyle = '--', alpha = 0.3)

## 3.4 添加标题
plt.title('北京11点~12点的气温变化折线图', fontsize = 20)

## 3.5 添加标签
plt.xlabel('时间', fontsize = 12)
plt.ylabel('气温', fontsize = 12)

## 3.6保存图像
plt.savefig('.\\my_first_figure.png', dpi = 100)

# 4.显示图像
plt.show()
```
