# 1.concat()

- pd.concat([data1, data2, data3, ...], jion, axis)
- 既可以行合并，也可以列合并 -> 通过axis实现
- 可以合并多个数据 ( >= 2)

- jion ：  'outer' / 'inner' -> 满外合并 / 内合并
- axis ：  0 / 1 -> 列合并 / 行合并
- jion默认是满外合并（即jion = 'outer'）
- axis默认是列合并（即axis = 0）

- 满外合并：左边全集 + 右边全集 + 交集
- 内合并：交集

两种合并方式的对象是 索引值

-> 当两个df对象的索引不一样时，此时用'outer'， 则会有先将交集部分合并，然后两个各不相同的部分在依次放在下面，对应缺失的部分用nan来表示

-> 此时用'inner',则只合并交集部分，其余不同的部分全部舍去

```python
# 1.1 列合并
# 列合并
new_df1 = pd.concat([df1, df2]) # 列合并
new_df2 = pd.concat([df1, df2], axis=0) # 效果同上
new_df1
```

```python
# 1.2 行合并
new_df3 = pd.concat([df1, df2], axis=1) # 行合并
new_df3
```

```python
index_name = [0, 1, 2, 5, 6]
df2.index = index_name
df2
```

```python
# 1.3 行合并 + 满外合并
new_df4 = pd.concat([df1, df2], join = 'outer', axis=1)
new_df4
```

```python
# 行合并 + 内合并
new_df5 = pd.concat([df1, df2], join='inner', axis=1)
new_df5
```

# 2.merge()

- pd.merge(left, right, how, on)
- 只可以行合并
- 只可以合并两个数据表

- left ： 左边的数据表
- right ： 右边的数据表
- how ： 合并方式 -> 'outer' / 'inner' / 'left' / 'right'
- on ： 指定的合并字段（列名）

```python
new_df6 = pd.merge(df1, df2, how='outer', on = ['City Name', 'Package'])
new_df7 = pd.merge(df1, df2, how='left', on = ['City Name', 'Package'])
new_df8 = pd.merge(df1, df2, how='right', on = ['City Name', 'Package'])
new_df9 = pd.merge(df1, df2, how='inner', on = ['City Name', 'Package'])
new_df6
```
