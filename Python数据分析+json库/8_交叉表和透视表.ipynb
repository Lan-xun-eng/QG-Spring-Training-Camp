观察下面可见：透视表是对分组聚合的进一步优化，即在将数据集以表格的形式显示出来，更好地表示某一列与另一列的关系

```python
df.groupby(['city', 'channel']).agg({
    'revenue':'sum'
})
```

```python
df.pivot_table(index = 'city', columns = 'channel', values = 'revenue', aggfunc = 'sum')
```
