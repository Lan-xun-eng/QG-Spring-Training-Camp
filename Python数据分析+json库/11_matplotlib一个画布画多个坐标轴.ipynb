```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir(r'E:\\pycharm项目\\data_analysis')

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 准备数据
x = range(0, 60)
y_beijing = [np.random.uniform(0, 15) for i in range(0, 60)]
y_shanghai = [np.random.uniform(15, 28) for i in range(0, 60)]

# 创建画布
fig, axes = plt.subplots(nrows = 1,ncols = 2, figsize = (25, 10), dpi = 100)

# 绘制初始折线图
axes[0].plot(x, y_beijing, label = '北京', color = 'c')
axes[1].plot(x, y_shanghai, label = '上海', color = 'pink')

# 添加刻度和刻度信息
x_ticks = range(0, 60, 5)
y_ticks = range(0, 40, 5)
x_ticks_label = [f'11点{i}分' for i in x_ticks]

axes[0].set_xticks(x_ticks, x_ticks_label)
axes[0].set_yticks(y_ticks)

axes[1].set_xticks(x_ticks, x_ticks_label)
axes[1].set_yticks(y_ticks)

# 添加网格
axes[0].grid(linestyle = '--', alpha = 0.3)
axes[1].grid(linestyle = '--', alpha = 0.3)

# 添加标题
axes[0].set_title('北京中午11点~12点气温变化图', fontsize = 20)
axes[1].set_title('上海中午11点~12点气温变化图', fontsize = 20)

# 添加标签
axes[0].set_xlabel('时间', fontsize = 15)
axes[0].set_ylabel('气温', fontsize = 15)

axes[1].set_xlabel('时间', fontsize = 15)
axes[1].set_ylabel('气温', fontsize = 15)

# 添加图例
axes[0].legend(loc = 'best', fontsize = 15)
axes[1].legend(loc = 'best', fontsize = 15)

# 保存图像
plt.savefig('.\\my_third_picture.png', dpi = 100)

# 显示图像
plt.show()
```
