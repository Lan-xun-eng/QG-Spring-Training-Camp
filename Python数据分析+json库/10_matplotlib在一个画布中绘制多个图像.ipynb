```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir(r'E:\\pycharm项目\\data_analysis')

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 准备数据
x = range(0, 60)
y_beijing = [np.random.uniform(0, 15) for i in range(0, 60)]
y_shanghai = [np.random.uniform(15, 28) for i in range(0, 60)]

# 创建画布
plt.figure(figsize = (20, 8), dpi = 100)

# 绘制初始折线图
plt.plot(x, y_beijing, label = '北京', color = 'c')
plt.plot(x, y_shanghai, label = '上海', color = 'pink')

# 添加刻度和刻度信息
x_ticks = range(0, 60, 5)
y_ticks = range(0, 40, 5)
x_ticks_label = [f'11点{i}分' for i in x_ticks]

plt.xticks(x_ticks, x_ticks_label, fontsize = 12)
plt.yticks(y_ticks, fontsize = 12)

# 添加网格
plt.grid(linestyle = '--', alpha = 0.3)

# 添加标题
plt.title('北京和上海11点~12点气温变化图', fontsize = 20)

# 添加标签
plt.xlabel('时间', fontsize = 15)
plt.ylabel('气温', fontsize = 15)

# 添加图例
plt.legend(loc = 'best', fontsize = 15)

# 保存图像
plt.savefig('.\\my_second_picture.png', dpi = 100)

# 显示图像
plt.show()
```
