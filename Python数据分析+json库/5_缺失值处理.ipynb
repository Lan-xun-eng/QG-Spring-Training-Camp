```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("E:\\pycharm项目\\data_analysis\\data\\US-pumpkins_data.xlsx")
df = df.drop(['Type'], axis=1)
df
```

# 1.判断是否含有缺失值

```python
# 1.1 isnull()
# pd.isnull(df对象) -> 是否为空（缺失值）
pd.isnull(df)
np.all(pd.isnull(df))
```

```python
# 1.2 notnull()
# pd.notnull() -> 判断是否为非空（非缺失值）
pd.notnull(df)
np.all(pd.notnull(df))
```

# 2. 删除缺失值

- df对象.dropna()
- 默认是删除行 -> 如果要删除列，则加入axis参数，即axis = 1
- 不改变原数据 -> 如果要改变原数据，则要加入inplace = True

```python
# 查看数据介绍，从中确定是否有缺失值
df.columns
df.info()
df.describe()
```

```python
# 删除缺失值
df.dropna()
```

# 3.替换缺失值

- df对象.fillna()
- 一般不会改变原数据 -> 除非加入 inplace = True ,原数据改变

```python
# 替换缺失值，往往使用for循环
# 先 找出每一列，再 判断每一列是否有缺失值，最后 将有缺失值的位置替换成对应的数据（此处全部替换成'a'）
for col_name in df.columns:
    # np.all()类似于逻辑与 -> True：该列无缺失值，False：该列存在缺失值
    # pd.notnull(df[col_name])  判断该列（的某个值）是否为非缺失值
    if np.all(pd.notnull(df[col_name])) == False:
        df[col_name].fillna('a', inplace = True)
```

# 4.转换为缺失值

- 背景：在实际开发中，并不是所有的缺失值都是用nan来表示的，可能会用 ？ 或者 空格 来表示，那么要如何删除或替换这些缺失值呢？
- 思路：？ -> nan -> 删除/替换

-       df对象.replace(to_replace, value)
        '''
        to_replace -> 被替换的值
        value -> 用于替换的值
        '''

```python
wis = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data")
wis = wis.head(30)
wis
```

```python
wis.replace(to_replace = '?', value = np.nan, inplace = True)
wis
```

```python
for col_name in wis.columns:
    if np.all(pd.notnull(wis[col_name])) == False:
        wis[col_name].fillna('a', inplace = True)
wis
```
