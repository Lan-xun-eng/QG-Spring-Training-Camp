# 1.分组聚合

-       df对象.groupby(['分组字段1', '分组字段2', ...]).get_group(('列名1', '列名2', ...))

！！！ 细节1：group()中是用[]包裹，而get_group()中是用()包裹

！！！ 细节2：前面group()中有多少个列名，则后面get_group()中就要提供多少信息

-       df对象.groupby(['分组字段1', '分组字段2', ...]).agg({
            '列名1':'函数名1',
            '列名2':'函数名2',
            ...
            })

！！！ 细节1：函数名 -> 不加()

！！！ 细节2：可以是相同草组0哦，也可以是不同操作

```python
# 1.1 按照单列分组
df.groupby('city')
```

```python
# 按照两列分组
df.groupby(['city', 'channel'])
```

```python
# 获得分组后某个组的数据
df.groupby(['city', 'channel']).get_group(('北京', '线下'))
```

```python
# 分组 + 聚合
# 场景1：通过 城市 和 销售渠道 分组，对 销售金额 计算其总和
df.groupby(['city', 'channel']).agg({
    'revenue':'sum'
    })

# 场景2：通过 城市 和 销售渠道 分组，分别对 销售金额计算其总和 + 成本计算其平均值
df.groupby(['city', 'channel']).agg({
    'revenue':'sum',
    'unit_cost':'mean'
    })
```

# 2. 分组过滤

-       df对象.groupby(['分组字段1', '分组字段2', ...]).filter(
            lambda x: dosomething return True/False
            )

！！！ 细节1：此处x指的是除了 城市 和 销售渠道 的其他全部列

```python
# 2.1 分组 + 过滤
# 场景：通过 城市 ，过滤出 销售金额 > 200 的数据
df.groupby('city').filter(lambda x : x['revenue'].mean() > 200)
# df.groupby('city').filter(lambda x : x.revenue.mean() > 200).city.drop_duplicates()
# df.groupby('city').filter(lambda x : x.revenue.mean() > 200).city.unique()
```
