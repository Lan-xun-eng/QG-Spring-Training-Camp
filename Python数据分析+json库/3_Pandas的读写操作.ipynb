```python
import pandas as pd
import numpy as np
```

# 1.csv文件

```python
# 1.1 读取csv文件
# read_csv(文件路径, sep, usecols, index_col)
'''
sep 分割符，默认是','
usecols 选择要读取的列名；若不选，则默认去全读
index_col 选择要作为索引的列名
'''
df1 = pd.read_csv("E:\\pycharm项目\\data_analysis\\pumpkins_data.csv",
                 # sep = ',',
                 # usecols = ['City Name', 'Type', 'Package', 'Low Price']
                  )
df1
```

```python
# 1.2 写出csv文件
# to_csv(文件路径, sep, usecols, index_col, index,  header, mode, encoding)
'''
sep 分隔符
usecols 要读出的列名；若没写，则全部读出
index True/False 表示是否要读出索引
header True/False 表示是否要读出表头，即列名 -> 一般是True
mode 'w'/'a'
encoding 码表
'''
df1.to_csv("E:\\pycharm项目\\data_analysis\\my_file1.csv", sep = ',', index = False, header = True, mode = 'w')
```

```python
# 1.3 特殊的csv文件：tsv
# 区别：csv -> 以','为分隔符   tsv -> 以'\t'（即Tab键）为分隔符

# 读取tsv文件
df2 = pd.read_csv("E:\\pycharm项目\\data_analysis\\tsv演示文件.txt", sep = '\t')
df2

# 写出tsv文件
df2.to_csv("E:\\pycharm项目\\data_analysis\\my_file2.tsv", sep = '\t', index = False, header = True, mode = 'w')
```

# 3.JSON文件

```python
# 3.1 读取JSON文件
# read_json(文件路径, orient, lines)
'''
orient 读取的形式 -> 有：split, index, records, columns(默认的形式)
lines True/False 是否要逐行读取
'''
df3 = pd.read_json('./',orient = 'records', lines = True)
```

```python
# 3.2 写出json文件
# df对象.to_json(文件路径, orient, lines)
'''
orient 储存的json形式
lines 一个对象写成一行
'''
df3.to_json('./', orient = 'records', lines = True)
```
