```python
df1 = pd.read_excel("E:\\pycharm项目\\data_analysis\\data\\US-pumpkins_data.xlsx",
                   usecols = ['City Name', 'Package', 'Low Price', 'High Price']
                   )
df = df1[:5].copy()
df
```

# 1.增

```python
# 直接赋值
df['c1'] = 23  # 方式1：固定值
df['c2'] = ['a', 'b', 'c', 'd', 'e']  # 方式2：列表
df['c3'] = df['Low Price'] + 2  # 方式3：通过已有的列（Series）来计算

def func():
    return '2026-01-12'
df['c4'] = func() # 方式4：函数
df
```

```python
# assign()函数
# 新增1列
df.assign(c1 = 23)

# 新增n列
df.assign(
    c1 = 23,
    c2 = ['a', 'b', 'c', 'd', 'e']
)
```

# 2.删和去重

## 2.1 删

```python
# 2.1.1 drop()函数
# 既可以删除行，也可以删除列 -> 但是默认是删除行的

# 删除行
df.drop([0])
df.drop([0, 3])

# 删除列
# 删除列要加一个axis参数 -> axis = 1
df.drop('Package', axis = 1)
df.drop(['Package', 'Low Price'], axis = 1)
```

!!! drop()函数的删除是暂时的，如若希望其删除原始数据，则需加入inplace参数 -> inplace = True时，则删除的是原始数据

```python
# 2.1.2 del关键字
# 其只可以删除列
del df['Low Price']
df
```

！！！ del是直接删除原始数据的
